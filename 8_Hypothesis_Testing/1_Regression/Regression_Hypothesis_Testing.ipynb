# `SISSO Formula`

```python
import numpy as np
# Coefficients and constant c0 for the model
c1 = 0.1633392287E+00
c2 = 0.6385917246E-04
c3 = -0.2889033757E+02
c0 = -0.1911288617E+01

# Assuming df1 contains the required features:
# 'vand_rad_Min', 'Volume', 'atom_rad_Min', 'density_Avg', 'density_Min',
# 'atom_rad_Avg', 'vand_rad_Avg', 'cov_rad_Max', 'atom_rad_Max'

# Calculate the descriptors
df1['d001_No_Phase'] = np.exp(df1['vand_rad_Min'] / df1['Volume']) * (df1['atom_rad_Min'] / (df1['density_Avg'] + df1['density_Min']))

df1['d002_No_Phase'] = ((df1['atom_rad_Avg']**2 - df1['atom_rad_Min'] * df1['vand_rad_Avg']) / np.log(df1['density_Min']))

df1['d003_No_Phase'] = np.exp(-df1['density_Avg']) / ((df1['atom_rad_Avg'] - df1['cov_rad_Max']) + (df1['atom_rad_Max'] - df1['cov_rad_Max']))

df1 = df1.round(2)
```

# `p-test`

```python
from scipy import stats

descriptors = ['d001_No_Phase', 'd002_No_Phase', 'd003_No_Phase']

for desc in descriptors:
    # Drop NaN values
    clean = df1_No_Phase[[desc, 'BandGap']].dropna()

    # Pearson correlation and p-value
    corr, p_val = stats.pearsonr(clean[desc], clean['BandGap'])

    print(f"\nDescriptor: {desc}")
    print(f"  Pearson Correlation : {corr:.4f}")
    print(f"  P-value             : {p_val:.6e}")

    if p_val < 0.05:
        print(f"  Result:  Significant relationship with BandGap")
    else:
        print(f"  Result:  No significant relationship with BandGap")
```

# `T-test per Coefficient + F-test for Regression`

```python
from sklearn.preprocessing import StandardScaler
import statsmodels.api as sm
import pandas as pd

# Redefine clean explicitly
clean = df1_No_Phase[['d001_No_Phase', 'd002_No_Phase',
                       'd003_No_Phase', 'BandGap']].dropna()

X_raw = clean[['d001_No_Phase', 'd002_No_Phase', 'd003_No_Phase']]
y     = clean['BandGap']

# Apply StandardScaler
scaler   = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X_raw),
                         columns=['d001_No_Phase', 'd002_No_Phase', 'd003_No_Phase'])

# Add constant and fit
X_scaled = sm.add_constant(X_scaled)
model_scaled = sm.OLS(y, X_scaled).fit()
print(model_scaled.summary())
```

#  <mark>Hypothesis Testing for Phase Data</mark>

```python
# Step 1: Compute the 3D descriptors
df1['d001_Phase'] = (df1['atom_rad_Min'] / (df1['density_Avg'] + df1['density_Min'])) / \
              np.sqrt(df1['Volume'] / df1['vand_rad_Min'])

df1['d002_Phase'] = (df1['cov_rad_Min'] - df1['vand_rad_Avg']) * \
              np.abs((df1['atom_rad_Avg'] / df1['atom_rad_Min']) -
                     (df1['vand_rad_Avg'] / df1['atom_rad_Avg']))

df1['d003_Phase'] = (df1['ST'] * (df1['density_Avg'] - df1['density_Min'])) / \
              np.abs(df1['atom_rad_Min'] - (df1['vand_rad_Max'] - df1['cov_rad_Min']))
```

```python
from scipy import stats

descriptors = ['d001_Phase', 'd002_Phase', 'd003_Phase']

for desc in descriptors:
    # Drop NaN values
    clean = df1_Phase[[desc, 'BandGap']].dropna()

    # Pearson correlation and p-value
    corr, p_val = stats.pearsonr(clean[desc], clean['BandGap'])

    print(f"\nDescriptor: {desc}")
    print(f"  Pearson Correlation : {corr:.4f}")
    print(f"  P-value             : {p_val:.6e}")

    if p_val < 0.05:
        print(f"  Result:  Significant relationship with BandGap")
    else:
        print(f"  Result:  No significant relationship with BandGap")
```
