# `Hypothesis Testing After Data Synthesized`

```python
from scipy import stats

bandgap_1 = df[df['BandGap_class'] == 1]
bandgap_0 = df[df['BandGap_class'] == 0]

# Mean and standard deviation
print("Mean and Std for Bandgap = 1")
print(df[['Volume', 'density_Avg', 'atom_rad_Avg']].mean(), df[['Volume', 'density_Avg', 'atom_rad_Avg']].std())

print("Mean and Std for Bandgap = 0")
print(df[['Volume', 'density_Avg', 'atom_rad_Avg']].mean(), df[['Volume', 'density_Avg', 'atom_rad_Avg']].std())

# t-test for Volume * density_Avg
t_stat, p_val = stats.ttest_ind(bandgap_1['Volume'] * bandgap_1['density_Avg'], bandgap_0['Volume'] * bandgap_0['density_Avg'])
print(f'T-test for Volume * density_Avg: t_stat = {t_stat}, p_val = {p_val}')

# t-test for Volume * atom_rad_Avg
t_stat, p_val = stats.ttest_ind(bandgap_1['Volume'] * bandgap_1['atom_rad_Avg'], bandgap_0['Volume'] * bandgap_0['atom_rad_Avg'])
print(f'T-test for Volume * atom_rad_Avg: t_stat = {t_stat}, p_val = {p_val}')
```

# `Design Feature via SISSO`

```python
Actual_Data['Volume_density_Avg_cbrt']= (np.cbrt(Actual_Data['Volume']*Actual_Data['density_Avg'])).round(decimals = 2)
Actual_Data['Volume_atom_rad_Avg_cbrt']= (np.cbrt(Actual_Data['Volume']*Actual_Data['atom_rad_Avg'])).round(decimals = 2)
Actual_Data.head()
```

# `Use Logistic Regression Formula`

```python
Actual_Data1= Actual_Data[['compound', 'BandGap','BandGap_class', 'Volume',	'density_Avg','atom_rad_Avg','Volume_density_Avg_cbrt', 'Volume_atom_rad_Avg_cbrt']]
Actual_Data1.head(2)
```

```python
Actual_Data1.shape
```

```python
import numpy as np

# Define the logistic regression parameters
intercept = 59.7791
# Replace X1 and X2 with the corresponding coefficients for:
# c1 = 'Volume_density_Avg' and c2 = 'Volume_atom_rad_Avg'
lasso_coefficients = {
    'Volume_density_Avg_cbrt': -1.3922,
    'Volume_atom_rad_Avg_cbrt': -1.6524
}

# Compute logit(P) for each row in ML_data2
def classify(row):
    # Start with the intercept
    logit_P = intercept

    # Add the weighted contribution from each feature
    logit_P += lasso_coefficients['Volume_density_Avg_cbrt'] * row['Volume_density_Avg_cbrt']
    logit_P += lasso_coefficients['Volume_atom_rad_Avg_cbrt'] * row['Volume_atom_rad_Avg_cbrt']

    # Convert logit to probability using the sigmoid function
    probability = 1 / (1 + np.exp(-logit_P))

    # Assign class based on probability threshold of 0.5
    return 1 if probability > 0.5 else 0

# Example usage:
# Assume X1 is your DataFrame containing 'Volume_density_Avg' and 'Volume_atom_rad_Avg'
Actual_Data1['Predicted_Class'] = Actual_Data1.apply(classify, axis=1)
```

```python
Actual_Data1.head()
```

```python
from scipy import stats

bandgap_1 = Actual_Data1[Actual_Data1['Predicted_Class'] == 1]
bandgap_0 = Actual_Data1[Actual_Data1['Predicted_Class'] == 0]

# Mean and standard deviation
print("Mean and Std for Bandgap = 1")
print(bandgap_1[['Volume', 'density_Avg', 'atom_rad_Avg']].mean())
print(bandgap_1[['Volume', 'density_Avg', 'atom_rad_Avg']].std())

print("\nMean and Std for Bandgap = 0")
print(bandgap_0[['Volume', 'density_Avg', 'atom_rad_Avg']].mean())
print(bandgap_0[['Volume', 'density_Avg', 'atom_rad_Avg']].std())

# t-test for Volume * density_Avg
t_stat, p_val = stats.ttest_ind(bandgap_1['Volume'] * bandgap_1['density_Avg'], bandgap_0['Volume'] * bandgap_0['density_Avg'])
print(f'\nT-test for Volume * density_Avg: t_stat = {t_stat}, p_val = {p_val}')

# t-test for Volume * atom_rad_Avg
t_stat, p_val = stats.ttest_ind(bandgap_1['Volume'] * bandgap_1['atom_rad_Avg'], bandgap_0['Volume'] * bandgap_0['atom_rad_Avg'])
print(f'T-test for Volume * atom_rad_Avg: t_stat = {t_stat}, p_val = {p_val}')
```

# `T-Test Results`
Volume × Density (p = 6.03e-07): The result is highly significant. The negative t-statistic (-5.44) confirms that this combined feature is substantially lower in bandgap materials. The difference is not due to chance.
Volume × Atomic Radius (p = 1.60e-08): This is even more significant. With a t-statistic of -6.31, atomic radius combined with volume is the stronger separator between the two classes. Bandgap materials have a much smaller volume-radius product.
