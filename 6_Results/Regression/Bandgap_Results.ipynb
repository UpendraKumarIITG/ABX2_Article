```python
##Import the modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
pd.set_option('display.max_columns', 500)
#df = pd.read_csv('Descriptor_GaBP2__20_Sep_24.csv')
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/GaBP2/GaBP2_2024/4_SISSO_Phase_Electronegativity/Descriptor_GaBP2__20_Sep_24.csv')
df.head(2)
```

```python
df1 = df.copy()
```

# `SISSO Formula`

```python
import numpy as np

# Coefficients and constant c0 for the model
c1 = 0.1633392287E+00
c2 = 0.6385917246E-04
c3 = -0.2889033757E+02
c0 = -0.1911288617E+01

# Assuming df1 contains the required features:
# 'vand_rad_Min', 'Volume', 'atom_rad_Min', 'density_Avg', 'density_Min',
# 'atom_rad_Avg', 'vand_rad_Avg', 'cov_rad_Max', 'atom_rad_Max'

# Calculate the descriptors
df1['d001_No_Phase'] = np.exp(df1['vand_rad_Min'] / df1['Volume']) * (df1['atom_rad_Min'] / (df1['density_Avg'] + df1['density_Min']))

df1['d002_No_Phase'] = ((df1['atom_rad_Avg']**2 - df1['atom_rad_Min'] * df1['vand_rad_Avg']) / np.log(df1['density_Min']))

df1['d003_No_Phase'] = np.exp(-df1['density_Avg']) / ((df1['atom_rad_Avg'] - df1['cov_rad_Max']) + (df1['atom_rad_Max'] - df1['cov_rad_Max']))

df1 = df1.round(2)
```

```python
df1.head()
```

# `LASSO Bandgap Formula`

```python
df1['LASSO_Bandgap'] = 0.16*df1['d001_No_Phase'] +0.000060*df1['d002_No_Phase']-26.82*df1['d003_No_Phase']-1.92
df1 = df1.round(2)
df1.head()
```

```python
rmse_test = (np.sqrt(mean_squared_error(df1['BandGap'], df1['LASSO_Bandgap'])))
rmse_test = round(rmse_test,2)
r2_test = r2_score(df1['BandGap'], df1['LASSO_Bandgap'])
r2_test = round(r2_test,2)
print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse_test))
print('R$^2$ score is {}'.format(r2_test))
```

```python
# The model performance for testing set
# --------------------------------------
# RMSE is 0.32
# R$^2$ score is 0.92
```

# `LASSO Bandgap Formula Using Phase Features`

```python
# Step 1: Compute the 3D descriptors
df1['d001_Phase'] = (df1['atom_rad_Min'] / (df1['density_Avg'] + df1['density_Min'])) / \
              np.sqrt(df1['Volume'] / df1['vand_rad_Min'])

df1['d002_Phase'] = (df1['cov_rad_Min'] - df1['vand_rad_Avg']) * \
              np.abs((df1['atom_rad_Avg'] / df1['atom_rad_Min']) -
                     (df1['vand_rad_Avg'] / df1['atom_rad_Avg']))

df1['d003_Phase'] = (df1['ST'] * (df1['density_Avg'] - df1['density_Min'])) / \
              np.abs(df1['atom_rad_Min'] - (df1['vand_rad_Max'] - df1['cov_rad_Min']))
```

```python
df1['LASSO_Bandgap_Phase'] = 0.397959*(df1['d001_Phase']) +0.032042*(df1['d002_Phase'])-1.406245*df1['d003_Phase']-1.57
df1 = df1.round(decimals = 2)
df1.head()
```

```python
rmse_test = (np.sqrt(mean_squared_error(df1['BandGap'], df1['LASSO_Bandgap_Phase'])))
rmse_test = round(rmse_test,2)
r2_test = r2_score(df1['BandGap'], df1['LASSO_Bandgap_Phase'])
r2_test = round(r2_test,2)
print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse_test))
print('R$^2$ score is {}'.format(r2_test))
```

```python
# The model performance for testing set
# --------------------------------------
# RMSE is 0.24
# R$^2$ score is 0.96
```
